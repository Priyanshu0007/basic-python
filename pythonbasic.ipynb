```python
import cv2
df=cv2.CascadeClassifier(cv2.data.haarcascades+"haarcascade_frontalFace_default.xml")
vid=cv2.VideoCapture(0)
flag2=True
while flag2:
    flag,img=vid.read()
    img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    if flag:
        faces=df.detectMultiScale(img,1.1,5)
        # to detect face using rect and circle 
        for x,y,w,h in faces:
            img=cv2.rectangle(img,pt1=(x,y),pt2=(x+w,y+h),color=(0,0,255))
            img=cv2.circle(img, center=(x+w//2,y+w//2), radius=(5+w//2), color=(0,0,255))
        cv2.imshow("webcam",img)
        key=cv2.waitKey(1)
        if key==ord('q'):
            break
    else:
        break
cv2.destroyAllWindows()
```
